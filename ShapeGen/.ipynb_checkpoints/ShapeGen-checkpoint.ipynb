```python
import matplotlib.pyplot as plt
import os, random
from random import randint
from PIL import Image,ImageDraw

def PolygonArea(corners):
    n = len(corners) # of corners
    area = 0.0
    for i in range(n):
        j = (i + 1) % n
        area += corners[i][0] * corners[j][1]
        area -= corners[j][0] * corners[i][1]
    area = abs(area) / 2.0
    return area

def ccw(A,B,C):
    return (C[1]-A[1]) * (B[0]-A[0]) > (B[1]-A[1]) * (C[0]-A[0])

# Return true if line segments AB and CD intersect
def intersect(A,B,C,D):
    return ccw(A,C,D) != ccw(B,C,D) and ccw(A,B,C) != ccw(A,B,D)

def genBg():
    if bgimg:
        file = random.choice(os.listdir("backgrounds"))
        im = Image.open("backgrounds/"+file)
        p1, p2 = im.size
        p1 = randint(0,p1-size)
        p2 = randint(0,p2-size)
        area = (p1,p2,p1+size,p2+size)
        im = im.crop(area)
    else: im = Image.new('RGB', (size, size),bColor)
    return im

#Variables that affect the generation of the samples
buffer = 3
size = 40
minSize = 10
generateImages = True
randomizeLineThickness = True
fillMode = 2 # 1=outline 2 = fill 3 = random
bgimg = False
randClr = True
clrDist = 40

path = ''
path += 'Fill'+ str(fillMode)
if(not bgimg):
    path += 'No'
path += 'Bg'

if generateImages:
    imageCount = 20000
else:
    imageCount = 20
if randomizeLineThickness:
    def rnd():
        return randint(1,5)
else:
    def rnd():
        return randint(1,1)
    
if fillMode == 1:
    def fill():
        return False
if fillMode == 2:
    def fill():
        return True
if fillMode == 3:
    def fill():
        return (False, True)[randint(0,1)] #ternary operator returns False if 0(False) and True if 1(True)

if randClr:
    def getClr():
        return randint(0,255)
else:
    def getClr():
        return 0
```

```python
i=imageCount
plt.figure(figsize=(20,10))
while(i>0):
    i-=1

    lColor = getClr()
    bColor = randint(0,255-clrDist)
    if(bColor >= lColor - clrDist):
        bColor += clrDist*2
    lColor = (lColor, lColor, lColor)
    bColor = (bColor, bColor, bColor)
    
    im = genBg()

    # Generate 4 points inside image
    while True:
        A = (randint(buffer, size-buffer),randint(buffer, size-buffer))
        B = (randint(buffer, size-buffer),randint(buffer, size-buffer))
        C = (randint(buffer, size-buffer),randint(buffer, size-buffer))
        D = (randint(buffer, size-buffer),randint(buffer, size-buffer))
        # Make sure points are not even close to being collinear
        if(PolygonArea([A,B,C])>size
        and PolygonArea([A,B,D])>size
        and PolygonArea([A,C,D])>size
        and PolygonArea([B,C,D])>size):
            break

    draw = ImageDraw.Draw(im)
    if fill():
        pc = 0
        if(intersect(A, B, C, D)):
            draw.polygon([A, B, C], fill = lColor)
            draw.polygon([A, B, D], fill = lColor)
            pc += 1
        if(intersect(A, C, B, D)):
            draw.polygon([A, C, B], fill = lColor)
            draw.polygon([A, C, D], fill = lColor)
            pc += 1
        if (pc < 2 or intersect(A, D, C, B)):
            draw.polygon([A, D, B], fill = lColor)
            draw.polygon([A, D, C], fill = lColor)
    else:
        lc = 0 #line count
        lineWidth = rnd()
        if(not intersect(A, B, C, D)):
            draw.line(A+B,fill=lColor,width=lineWidth)
            draw.line(C+D,fill=lColor,width=lineWidth)
            lc+=1

        if(not intersect(B, C, A, D)):
            draw.line(B+C,fill=lColor,width=lineWidth)
            draw.line(A+D,fill=lColor,width=lineWidth)
            lc+=1

        if (lc < 2 and not intersect(A, C, B, D)):
            draw.line(A+C,fill=lColor,width=lineWidth)
            draw.line(B+D,fill=lColor,width=lineWidth)
    
    border = (min(A[0],B[0],C[0],D[0])-buffer,
              min(A[1],B[1],C[1],D[1])-buffer,
              max(A[0],B[0],C[0],D[0])+buffer,
              max(A[1],B[1],C[1],D[1])+buffer) # left, up, right, bottom
    im = im.crop(border)
    im = im.resize((size,size))

    if generateImages:
        im.save('../ShapeData/'+path+'/Quadrilateral/quad'+str(i)+'.png')
    else:
        plt.subplot(20/5 + 1, 10, 20-i)
        plt.imshow(im,cmap='gray',vmin=0,vmax=255)
        plt.axis('off')
```

```python
i=imageCount
plt.figure(figsize=(20,10))
while(i>0):
    i-=1

    lColor = getClr()
    bColor = randint(0,255-clrDist)
    if(bColor >= lColor - clrDist):
        bColor += clrDist*2

    lColor = (lColor, lColor, lColor)
    bColor = (bColor, bColor, bColor)
    
    im = genBg()

    #generate 3 points inside image
    while True:
        A = (randint(buffer, size - buffer), randint(buffer, size - buffer))
        B = (randint(buffer, size - buffer), randint(buffer, size - buffer))
        C = (randint(buffer, size - buffer), randint(buffer, size - buffer))
        if(PolygonArea([A,B,C])>size*2):
            break

    draw = ImageDraw.Draw(im)
    if fill():
        draw.polygon([A, B, C], fill = lColor)
    else:
        lineWidth = rnd()
        draw.line(A+B,fill=lColor,width=lineWidth)
        draw.line(A+C,fill=lColor,width=lineWidth)
        draw.line(B+C,fill=lColor,width=lineWidth)
    
    border = (min(A[0],B[0],C[0])-buffer,
              min(A[1],B[1],C[1])-buffer,
              max(A[0],B[0],C[0])+buffer,
              max(A[1],B[1],C[1])+buffer) # left, up, right, bottom
    im = im.crop(border)
    im = im.resize((size,size))
    
    if generateImages:
        im.save('../ShapeData/'+path+'/Triangle/Triangle'+str(i)+'.png')
    else:
        plt.subplot(20/5 + 1, 10, 20-i)
        plt.imshow(im,cmap='gray',vmin=0,vmax=255)
        plt.axis('off')
```

```python
i=imageCount
plt.figure(figsize=(20,10))
while(i>0):
    i-=1

    lColor = getClr()
    bColor = randint(0,255-clrDist)
    if(bColor >= lColor - clrDist):
        bColor += clrDist*2

    lColor = (lColor, lColor, lColor)
    bColor = (bColor, bColor, bColor)
    
    im = genBg()

    #generate 2 points inside image (topleft and bottomright)
    A = (randint(buffer, size - buffer - minSize), randint(buffer, size - buffer-minSize))
    B = (randint(A[0]+minSize, size - buffer), randint(A[1]+minSize, size - buffer))
    
    draw = ImageDraw.Draw(im)
    if fill():
        draw.ellipse(A+B, fill=lColor)
    else:
        draw.ellipse(A+B, outline=lColor, width = rnd())
        
    im = im.crop((A[0]-buffer,
                  A[1]-buffer,
                  B[0]+buffer,
                  B[1]+buffer))
    im = im.resize((size,size))
    
    if generateImages:
        im.save('../ShapeData/'+path+'/Ellipse/Ellipse'+str(i)+'.png')
    else:
        plt.subplot(20/5 + 1, 10, 20-i)
        plt.imshow(im,cmap='gray',vmin=0,vmax=255)
        plt.axis('off')
```
